```python
import numpy as np
import pandas as pd

file=open("../Data/corpus.txt")
data=file.read()
print(data)
```

```python
def remove_punctuations(data):
    punc='''!()-[]{};:'"\,<>./?@#$%^&*_~'''
    
    for i in punc:
        data=data.replace(i,'')
    
    return data

original_data=remove_punctuations(data).split(" ")
print(original_data)
data=list(set(original_data))
```

```python
df=pd.DataFrame(data,columns=['words'])
```

```python
def onehot(n,l):
    t=np.zeros((l,1))
    t[n]=1
    return t

def initialise_weights(TotalW,TotalB,a,b):
    W=np.zeros((a,b))+np.random.randn(a,b)
    B=np.zeros((b,1))+np.random.randn(b,1)
    TotalW.append(W)
    TotalB.append(B)

    return TotalW,TotalB

def normalise_scale(N,a,b):
    minN=min(N)
    maxN=max(N)
    for i in range(len(N)):
        N[i]=((N[i]-minN)/(maxN-minN))*(a-b)+b
    return N
    
def cost(true,pred):
    return -np.sum(true*np.log2(pred))
    
def softmax(z):
    return np.exp(z)/np.sum(np.exp(z))
    
def sigmoid(z):
    return 1/(1+np.exp(-z))

def minibatch_gradient_descent(TotalW,TotalB,data,target,l,max_iterations,learning_rate,batch_size):
    cnt=0
    flag=0
    while(cnt<max_iterations):
        cnt+=1
        
        TotalgradW=[]
        TotalgradB=[]

        for i in range(len(data)):  
            pred,TotalI=forward_propagation(TotalW,TotalB,data[i],l)

            c=cost(target[i],pred)
            
            gradW=backward_propagation(data[i],TotalW,target[i],pred,TotalI,l)
            
            if(i==0 or flag==1):
                TotalgradW=gradW
                #TotalgradB=gradB
                flag=0
            else:
                for j in range(len(gradW)):
                    TotalgradW[j]+=gradW[j]
                    #TotalgradB[j]+=gradB[j]
            
            if((i+1)%batch_size==0):

                for j in range(len(TotalW)):
                    TotalW[j]-=learning_rate*TotalgradW[j]
                    #TotalB[j]-=learning_rate*TotalgradB[j]
                flag=1
                    
                print("loss ---- ",c)
                
def forward_propagation(TotalW,TotalB,data,l):
    TotalI=[]
    TotalI.append(data)
    for i in range(l-1):
        data=(np.dot(TotalW[i].T,data)+TotalB[i])
        TotalI.insert(0,data)
    
    data=softmax(np.dot((TotalW[i+1].T),data)+TotalB[i+1])
    return data,TotalI
    
def findindex(d):
    for i in range(len(d)):
        if(d[i]==[1]):
            return i

def backward_propagation(d,TotalW,one_hot,pred,TotalI,l):
    TotalgradW=[]
    
    i=findindex(d)
    grad_final_layer=-(one_hot-pred)
    grad_hidden_layer=grad_final_layer
    
    gradW=np.dot(TotalI[0],grad_hidden_layer.T)
    TotalgradW.insert(0,gradW)

    val=np.dot(TotalW[1],grad_hidden_layer)
    gradW=np.zeros(TotalW[0].shape)
    gradW[i]+=val.reshape(val.shape[0])
    TotalgradW.insert(0,gradW)
    
    return TotalgradW
```

```python
TotalW=[]
TotalB=[]

data=[]
for i in range(len(df)):
    data.append(onehot(i,len(df)))
data=np.asarray(data)

d={}
for i in range(len(df)):
    d[df['words'][i]]=data[i]

target=[]
for i,j in d.items():
    index=original_data.index(i)
    if(index+1<len(original_data)):
        target.append(d[original_data[index+1]])
    else:
        target.append(d[original_data[index]])

input_neurons=len(data[0])
output_neurons=len(data[0])
max_iterations=500

#Parameter1
TotalW,TotalB=initialise_weights(TotalW,TotalB,input_neurons,10)
#Parameter2
TotalW,TotalB=initialise_weights(TotalW,TotalB,10,input_neurons)

#Total Layers
l=len(TotalW)

#learning rate
learning_rate=0.01

minibatch_gradient_descent(TotalW,TotalB,data,target,l,max_iterations,learning_rate,1)
```
